```
def acoustic_features(file_path):
  y, sr = librosa.load(file_path, sr = 22050)
  #features = []
  chroma_stft = lf.chroma_stft(y, sr)	#Compute a chromagram from a waveform or power spectrogram.
  chroma_stft_mean = [np.mean(i) for i in chroma_stft]
  #chroma_stft_delat

  chroma_cqt = lf.chroma_cqt(y, sr)	#Constant-Q chromagram
  chroma_cqt_mean = [np.mean(i) for i in chroma_cqt]

  chroma_cens = lf.chroma_cens(y, sr)	#Computes the chroma variant Chroma Energy Normalized (CENS), following [R674badebce0d-1].
  chroma_cens_mean = [np.mean(i) for i in chroma_cens]

  melspectrogram = lf.melspectrogram(y, sr)	#Compute a mel-scaled spectrogram.
  melspectrogram_mean = [np.mean(i) for i in melspectrogram]

  mfcc = lf.mfcc(y, sr)	#Mel-frequency cepstral coefficients (MFCCs)
  mfcc_mean = [np.mean(i) for i in mfcc]
  
  rms = lf.rms(y, frame_length = sr)	#Compute root-mean-square (RMS) value for each frame, either from the audio samples y or from a spectrogram S.
  rms_mean = [np.mean(i) for i in rms]
  
  spectral_centroid = lf.spectral_centroid(y, sr)	#Compute the spectral centroid.
  spectral_centroid_mean = [np.mean(i) for i in spectral_centroid]
  
  spectral_bandwidth = lf.spectral_bandwidth(y, sr)	#Compute pth-order spectral bandwidth.
  spectral_bandwidth_mean = [np.mean(i) for i in spectral_bandwidth]
  
  spectral_contrast = lf.spectral_contrast(y, sr)	#Compute spectral contrast [R6ffcc01153df-1]
  spectral_contrast_mean = [np.mean(i) for i in spectral_contrast]
  
  spectral_flatness = lf.spectral_flatness(y, n_fft = sr)	#Compute spectral flatness
  spectral_flatness_mean = [np.mean(i) for i in spectral_flatness]
  
  spectral_rolloff = lf.spectral_rolloff(y, sr)	#Compute roll-off frequency.
  spectral_rolloff_mean = [np.mean(i) for i in spectral_rolloff]
  
  poly_features = lf.poly_features(y, sr)	#Get coefficients of fitting an nth-order polynomial to the columns of a spectrogram.
  poly_features_mean = [np.mean(i) for i in poly_features]
  
  tonnetz = lf.tonnetz(y, sr)	#Computes the tonal centroid featurevs (tonnetz), following the method of [Recf246e5a035-1].
  tonnetz_mean = [np.mean(i) for i in tonnetz]
  
  zero_crossing_rate = lf.zero_crossing_rate(y, frame_length = 22050)
  zero_crossing_rate_mean = [np.mean(i) for i in zero_crossing_rate]
  
  #features.append(chroma_stft, chroma_cqt, chroma_cens, melspectrogram, mfcc, rms, spectral_centroid, spectral_bandwidth, spectral_contrast, spectral_flatness, spectral_rolloff, poly_features, tonnetz, zero_crossing_rate)
  output = chroma_stft_mean, chroma_cqt_mean, chroma_cens_mean, melspectrogram_mean, mfcc_mean, rms_mean, spectral_centroid_mean, spectral_bandwidth_mean, spectral_contrast_mean, spectral_flatness_mean, spectral_rolloff_mean, poly_features_mean, tonnetz_mean, zero_crossing_rate_mean
  output_columns_dict = {'chroma_stft_mean': chroma_stft_mean, 'chroma_cqt_mean': chroma_cqt_mean,'chroma_cens_mean': chroma_cens_mean, 'melspectrogram_mean': melspectrogram_mean,'mfcc_mean': mfcc_mean, 'rms_mean': rms_mean,'spectral_centroid_mean': spectral_centroid_mean, 'spectral_bandwidth_mean': spectral_bandwidth_mean,'spectral_contrast_mean': spectral_contrast_mean, 'spectral_flatness_mean': spectral_flatness_mean,'spectral_rolloff_mean': spectral_rolloff_mean, 'poly_features_mean': poly_features_mean,'tonnetz_mean': tonnetz_mean, 'zero_crossing_rate_mean': zero_crossing_rate_mean}
  flattened_output = [mean for list_means in output for mean in list_means]
  return flattened_output, output_columns_dict
```
